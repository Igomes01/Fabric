```python
df = spark.read.format("csv").option("header","true").load("abfss://63cc2aba-b9db-405b-84ab-d2d47b1478c2@onelake.dfs.fabric.microsoft.com/d158701a-0912-45e8-a140-4f06215a8400/Files/praia-copacabana-rj.csv")
# df now is a Spark DataFrame containing CSV data from "abfss://63cc2aba-b9db-405b-84ab-d2d47b1478c2@onelake.dfs.fabric.microsoft.com/d158701a-0912-45e8-a140-4f06215a8400/Files/praia-copacabana-rj.csv".
display(df)
```

```python
# Divide a coluna em tr√™s novas colunas: comentario, nota e data
import pandas as pd
from pyspark.sql.functions import split

df_dividido = df.withColumn("comentario", split("comentario;nota;data", ";").getItem(0)) \
                .withColumn("nota", split("comentario;nota;data", ";").getItem(1)) \
                .withColumn("data", split("comentario;nota;data", ";").getItem(2))

df_final = df_dividido.drop("comentario;nota;data")
```

```python
df_final.write.mode("overwrite").format("delta").saveAsTable("comentarios_copacabana")
```
